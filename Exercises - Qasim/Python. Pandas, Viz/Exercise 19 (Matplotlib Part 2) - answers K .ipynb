### Figure size, aspect ratio and DPI

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

For example:

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2

fig = plt.figure(figsize=(12,3), dpi=72)
ax1 = fig.add_subplot(111)
ax1.plot(x,y,'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('title')
```

### Colours
With matplotlib, we can define the colors of lines and other graphical elements in a number of ways. First of all, we can use the MATLAB-like syntax where `'b'` means blue, `'g'` means green, etc. The MATLAB API for selecting line styles are also supported: where, for example, 'b.-' means a blue line with dots:

```python
import matplotlib.pyplot as plt
import numpy as np

def list_square(a_list):
    return [element**2 for element in a_list]

def list_cube(a_list):
    return [element**3 for element in a_list]

x = np.arange(0,5,0.5)
y = list_square(x)
z = list_cube(x)

fig = plt.figure(figsize=(7.6,5.5))
ax = fig.add_subplot(111)
ax.plot(x,y, 'b.-', label='x**2')
ax.plot(x,z,'g--', label='x**3')
ax.legend()
```

We can also define colors by their names or RGB hex codes and optionally provide an alpha value using the `color` and `alpha` keyword arguments. Alpha indicates opacity.

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(0,5,0.5)
a = x+1
b = x+2
c = x+3

fig = plt.figure(figsize=(7.6,5.5))
ax = fig.add_subplot(111)
ax.plot(x,a, color='mediumpurple')
ax.plot(x,b,color='indigo')
ax.plot(x,c,color='tab:orange')
```

### Line and Marker styles
To change the line width, we can use the `linewidth` or `lw` keyword argument. The line style can be selected using the `linestyle` or `ls` keyword arguments:

```python
x = np.arange(0,5,0.5)
a = x+1
b = x+2
c = x+3
d = x+4
e = x+5
f = x+6
g = x+7
h = x+8
i = x+9
j = x+10
k = x+11
l = x+12
m = x+13
n = x+14
o = x+15
q = x+16

fig = plt.figure(figsize=(12,5.5))
ax = fig.add_subplot(111)
ax.plot(x,a,'r',linewidth=0.25)
ax.plot(x,b, 'r',linewidth=0.75)
ax.plot(x,c, 'r',linewidth=1)
ax.plot(x,d, 'r',linewidth=2)
ax.plot(x,e,'g',linewidth=3)
ax.plot(x,f,'g-.',linewidth=3)
ax.plot(x,g, 'g:',linewidth=3)
ax.plot(x,h,'k--',linewidth=2)
ax.plot(x,i,'b-|',linewidth=3)
ax.plot(x,j, 'b--.',linewidth=3)
ax.plot(x,k,'b-s',linewidth=3)
ax.plot(x,l,'b.-',linewidth=3)
ax.plot(x,m, 'm.-',linewidth=0.75)
ax.plot(x,n,'mo-',linewidth=0.75)
ax.plot(x,o,'m',marker='o',markerfacecolor='r',markersize=7,linewidth=0.75)
ax.plot(x,q,'m',marker='s',markeredgecolor='g',ms=10, markerfacecolor='yellow', markeredgewidth=3,linewidth=0.75)
```

### Plot range
We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges:

```python
import matplotlib.pyplot as plt
import numpy as np

def list_square(a_list):
    return [element**2 for element in a_list]

def list_cube(a_list):
    return [element**3 for element in a_list]

x = np.arange(0,5,0.10)
y = list_square(x)
z = list_cube(x)

fig1 = plt.figure(figsize=[12,4.5])
ax1 = fig1.add_subplot(131)
ax1.plot(x,y)
ax1.plot(x,z)
ax1.set_title('default axes ranges')

fig2 = plt.figure()
ax2 = fig1.add_subplot(132)
ax2.plot(x,y)
ax2.plot(x,z)
ax2.set_title('tight axes')
ax2.axis('tight')

fig3 = plt.figure()
ax3 = fig1.add_subplot(133)
ax3.plot(x,y)
ax3.plot(x,z)
ax3.set_xlim(2,5)
ax3.set_ylim(0,60)
ax3.set_title('custom axes range')
```
