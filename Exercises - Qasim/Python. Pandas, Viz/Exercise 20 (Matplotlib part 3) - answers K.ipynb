### Advanced Matplotlib Concepts

#### Logrithmic scale
It is also possible to set a logarithmic scale for one or both axes. This functionality is in fact only one application of a more general transformation system in Matplotlib. Each of the axes' scales are set seperately using `set_xscale` and `set_yscale` methods which accept one parameter (with the value "log" in this case):

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
import matplotlib.pyplot as plt


figure= plt.figure(figsize=(10, 4), dpi=72)
       
x = np.linspace(0, 5, 11)
y = x ** 2
z = np.exp(x)

axis1 = figure.add_subplot(1,2,1)  
axis1.plot(x,y)
axis1.plot(x,z)
axis1.set_title("Normal Scale")

axis2 = figure.add_subplot(1,2,2)    
axis2.plot(x,y)
axis2.plot(x,z)
axis2.set_yscale('log')
axis2.set_title(("Log Scale (y)"))

plt.show()
```

### Placement of ticks and custom ticks label
We can explicitly determine where we want the axis ticks with `set_xticks` and `set_yticks`, which both take a list of values for where on the axis the ticks are to be placed. We can also use the `set_xticklabels` and `set_yticklabels` methods to provide a list of custom text labels for each tick location:

```python
import matplotlib.pyplot as plt


figure= plt.figure(figsize=(10, 4), dpi=72)
       
x = np.linspace(0, 5, 11)
y = x ** 2
z = np.exp(x)

axis1 = figure.add_subplot(111)  
axis1.plot(x,y)
axis1.plot(x,z)

axis1.tick_params(labelsize=20,direction='out',pad=10)
axis1.set_xticklabels(['','','α', 'β', 'γ', 'δ', 'ε'])


# did not managed to remve the 1st tick mark
# issue with greek letters formatting as top is cropped
```

### Scientific Notation
With large numbers on axes, it is often better use scientific notation:

```python
import matplotlib.pyplot as plt
from matplotlib import ticker

figure= plt.figure(figsize=(6, 4), dpi=72)
       
x = np.linspace(0, 5, 11)
y = x ** 2
z = np.exp(x)


axis1 = figure.add_subplot(1,1,1)  
axis1.plot(x,y)
axis1.plot(x,z)


formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 

axis1.yaxis.set_major_formatter(formatter) 
axis1.set_title("scientific notation")
axis1.set_yticks([0, 50, 100, 150])
```
