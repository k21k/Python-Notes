# 1 Preparing the Data Set

```python
import numpy as np
import pandas as pd

df = pd.read_csv(r'C:\Users\delchain_default\Documents\GitHub\Kaggle-DataSets\Apple Mobility Data\applemobility.csv')
```

### Assessing the Data

```python
list(df.columns)
df.columns.astype(str)
```

```python
df_country = df[df['geo_type']=='country/region']
df_country
```

```python
df_country.drop(['geo_type','alternative_name','sub-region','country'], axis=1, inplace=True)
```

```python
df_country
```

### Creating 3 df for each transportation type, then transpose data

```python
df_country_driving = df_country[df_country.loc[:,'transportation_type'] == 'driving']
df_country_driving.drop(['transportation_type'], axis=1, inplace=True)

df_country_walking = df_country[df_country.loc[:,'transportation_type'] == 'walking']
df_country_walking.drop(['transportation_type'], axis=1, inplace=True)

df_country_transit = df_country[df_country.loc[:,'transportation_type'] == 'transit']
df_country_transit.drop(['transportation_type'], axis=1, inplace=True)
```

```python
df_country_driving = df_country_driving.transpose()
df_country_walking = df_country_walking.transpose()
df_country_transit = df_country_transit.transpose()
```

```python
df_country_driving.columns = df_country_driving.iloc[0]
df_country_driving.drop('region',inplace=True)
aa = df_country_driving
aa['transportation_type'] = 'driving'
```

```python
aa
```

```python
df_country_walking.columns = df_country_walking.iloc[0]
df_country_walking.drop('region', inplace=True)
bb = df_country_walking
bb['transportation_type'] = 'walking'
```

```python
df_country_transit.columns = df_country_transit.iloc[0]
df_country_transit.drop('region',inplace=True)
cc = df_country_transit
cc['transportation_type'] = 'transit'
```

### Creating the new df with the 3 df

```python
new_df = aa.append(bb).append(cc)
```

```python
# Create a new column with dates - emprirically, i found that important later for visualisation purposes

new_df.index =  pd.to_datetime(new_df.index, infer_datetime_format=True) 
new_df.index.name = 'time'
new_df = new_df.reset_index()

new_df['time'] = pd.to_datetime(new_df['time'], format = '%Y-%m-%d')
new_df['time'] = new_df['time'].strftime('%m/%d/%Y')

```

```python
new_df['time'].dtype
```

```python
new_df
```

### Filling NaN and empty values by baseline 100

```python
new_df.fillna(100, inplace=True)
new_df.replace("",100)
```

```python
new_df.to_csv('C:\Users\delchain_default\Documents\GitHub\Kaggle-DataSets\Apple Mobility Data\new_df.csv', index=True)
```
